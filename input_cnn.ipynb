```python
import cv2
import cv2 as cv
import os
import numpy as np

```

```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
```

```python
def get_snaps(image):
    
    # filtering
    blurred = cv.medianBlur(image, 5)
    gray = cv.cvtColor(blurred, cv.COLOR_BGR2GRAY)

    circles = cv.HoughCircles(gray, cv.HOUGH_GRADIENT, 1, 100,
                              param1=120, param2=35, minRadius=30, maxRadius=50).reshape(-1,3)

    # images of snaps

    detected_circles = np.uint16(np.around(circles)).reshape(-1,3)
    segment_size = 50

    snap_images = []
    for i in detected_circles:
        x = i[0]
        y = i[1]
        r = i[2]

        snap = image[y-segment_size:y+segment_size, x-segment_size:x+segment_size]
        snap_images.append(snap)

    snap_images = np.uint16(np.around(snap_images))#.reshape(2*segment_size,2*segment_size)
    
    return circles, snap_images
```

```python
#### WEIBLICH


# read in images

path1 = r'C:\Users\Fabian\OneDrive - rwth-aachen.de\Bachelorarbeit\Fotos DeltaT\27.10. ITA\w'
#path2 = ...
#path3 = ....
pathes = [path1]

images = []
images1 = [] # list of images
for path in pathes: 
    images_temp = load_images_from_folder(path)
    images_temp = np.asarray(images_temp)
    
    for i in range(images_temp.shape[0]):
        images.append(images_temp[i])

# saving snaps out of a list of images
images = images.copy()
snaps = []

for image in images:
    circles,snap_temp = get_snaps(image)
    for snap in snap_temp:
        snaps.append(snap)
        
        
# save snap images

for i in range(len(snaps)):
    cv.imwrite(r'C:\Users\Fabian\OneDrive - rwth-aachen.de\Bachelorarbeit\Fotos DeltaT\27.10. ITA/snaps/w/snap_w_%s.jpg' %(i+1), snaps[i])
```
